COMPARATIVO DE FAIXA ETARIA E GRAU DE ESCOLARIDADE DOS ELEITORES DA CIDADE DE SJC.

```python
import pandas as pd
import matplotlib.pyplot as plt
low_memory=False
plt.style.use('seaborn')
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
pd = pd.read_csv(r'C:\Users\Rita\perfil_eleitorado_2020.csv', sep = ";", error_bad_lines=False, encoding= "Latin 1")
fx=pd.query("NM_MUNICIPIO =='SÃO JOSÉ DOS CAMPOS'")
```

Breve descrição dos campos referente tabela utilizada para uso das informações 

DS_FAIXA_ETARIA = Descrição da faixa etaria a qual pertence o eleitor

NM_MUNICIPIO = Nome do município onde vota o eleitor 

QT_ELEITORES_PERFIL = Quantidade de Eleitores aptos a votar

DS_GRAU_ESCOLARIDADE = Descrição do grau de escolaridade declarada pelo eleitor

```python
fx.info()
```

```python
fx.sample(10)
```

```python
fx.groupby("DS_FAIXA_ETARIA").size().sort_values().tail(10).plot(kind='barh', title='Comparativo Idade dos Eleitores de SJC ')
```
